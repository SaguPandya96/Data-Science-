```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

##To check Data types

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
##Pie chart- top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation:- Zomato mazimum records or transaction are from india  and followed by USA , UK .Most business market captured in india.

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

##Observtions:-
1. when rating is between 4.5 to 4.9>> Excellent
2. When rating are between 4.0 to 3.4 >>Very good
3. when rating are btween 3.5 to 3.9 >>> Good
4. when rating is between 2.5 to 3.4 >>> average
5. when rating is between 1.8 to 2.4 >> Poor

```python
ratings.head()
```

```python
import matplotlib.pyplot as plt
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```
